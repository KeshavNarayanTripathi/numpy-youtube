NUMPY INDEXING AND SLICING OF MATRIX

```python
arr = np.arange(1, 31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
arr[0:2, 1:3]
```

```python
arr[3:6, 3:5]
```

```python
arr[:, 2]
```

```python
bool_index = arr%2 == 0
bool_index
```
